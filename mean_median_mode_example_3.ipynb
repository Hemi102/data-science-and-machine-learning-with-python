```python
import numpy as np
import matplotlib.pyplot as plt

students_low_var = np.random.normal(5000, 500, 1000)
students_high_var = np.random.normal(5000, 3000, 1000)


mean_low, std_low = np.mean(students_low_var), np.std(students_low_var)
mean_high, std_high = np.mean(students_high_var), np.std(students_high_var)

print("Low Variance Example -> Mean:", mean_low, "Std Dev:", std_low)
print("High Variance Example -> Mean:", mean_high, "Std Dev:", std_high)

plt.figure(figsize=(10,5))
plt.hist(students_low_var, bins=50, alpha=0.6, color="gray", label=f"Low Var (Std: {std_low:.0f})")
plt.hist(students_high_var, bins=50, alpha=0.6, color="red", label=f"High Var (Std: {std_high:.0f})")

plt.axvline(mean_low, color='blue', linestyle='dashed', linewidth=2, label="Low Var Mean")
plt.axvline(mean_high, color='red', linestyle='dashed', linewidth=2, label='High Var Mean')

plt.legend()
plt.title('Pocket Money Distribution (Low vs High Variance)')
plt.xlabel('Pocket Money (PKR)')
plt.ylabel('Number of Students')
plt.savefig('example3_hist.png', dpi=300, bbox_inches='tight')

plt.show()
```
