```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

incomes = np.random.normal(27000, 15000, 10000)
incomes =np.append(incomes,100000)
mean_val = np.mean(incomes)
median_val = np.median(incomes)
mode_val = stats.mode(stats.mode(incomes, keepdims = True))

print('Mean:', mean_val)
print('Median:', median_val)
print('Mode:', mode_val.mode[0], " (count:", mode_val.count[0], ")")

plt.hist(incomes, bins=50, alpha=0.3, color='blue')
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_val, color='green', linestyle='dashed', linewidth=2, label='Median')
plt.legend()
plt.show()
```
